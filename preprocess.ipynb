```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
```

## Tranformacoes base AG

```python
# Load originaal datasets
df_train_arit = pd.read_csv("databases/AG/Aritmética/TreinamentoDesbalanceadoArit.csv", sep=";", encoding="latin1")
df_train_esc = pd.read_csv("databases/AG/Escrita/TreinamentoDesbalanceadoEsc.csv", sep=";", encoding="latin1")
df_train_leit = pd.read_csv("databases/AG/Leitura/TreinamentoDesbalanceadoLeit.csv", sep=";", encoding="latin1")

df_valid_arit = pd.read_csv("databases/AG/Aritmética/TesteArit.csv", sep=";", encoding="latin1")
df_valid_esc = pd.read_csv("databases/AG/Escrita/TesteEsc.csv", sep=";", encoding="latin1")
df_valid_leit = pd.read_csv("databases/AG/Leitura/TesteLeit.csv", sep=";", encoding="latin1")

df_train_leit
```

```python
# Aritmetica
df_train_arit = df_train_arit.replace({'Sim': 1, 'Não': 0})
df_train_arit['TDE_MG_Esc'] = df_train_arit['TDE_MG_Esc'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_train_arit['Escolaridade_mae'] = df_train_arit['Escolaridade_mae'].replace({'EFundamental': 0, 'EMedio': 1, 'ESuperior':2})
df_train_arit['Parto'] = df_train_arit['Parto'].replace({'Normal': 0, 'Cesária': 1})

df_valid_arit = df_valid_arit.replace({'Sim': 1, 'Não': 0})
df_valid_arit['TDE_MG_Esc'] = df_valid_arit['TDE_MG_Esc'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_valid_arit['Escolaridade_mae'] = df_valid_arit['Escolaridade_mae'].replace({'EFundamental': 0, 'EMedio': 1, 'ESuperior':2})
df_valid_arit['Parto'] = df_valid_arit['Parto'].replace({'Normal': 0, 'Cesária': 1})

# Escrita
df_train_esc = df_train_esc.replace({'Sim': 1, 'Não': 0})
df_train_esc['TDE_MG_Leit'] = df_train_esc['TDE_MG_Leit'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_train_esc['TDE_MG_Arit'] = df_train_esc['TDE_MG_Arit'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_train_esc['Genero'] = df_train_esc['Genero'].replace({'Masculino': 0, 'Feminino': 1})
df_train_esc['Ano_escolar'] = df_train_esc['Ano_escolar'].replace({
    'EMedio': 9,
    'Nono': 8, 
    'Oitavo': 7,  
    'Sétimo': 6, 
    'Sexto': 5, 
    'Quinto': 4, 
    'Quarto': 3, 
    'Terceiro': 2,
    'Segundo': 1, 
    'Primeiro': 0})


df_valid_esc = df_valid_esc.replace({'Sim': 1, 'Não': 0})
df_valid_esc['TDE_MG_Leit'] = df_valid_esc['TDE_MG_Leit'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_valid_esc['TDE_MG_Arit'] = df_valid_esc['TDE_MG_Arit'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_valid_esc['Genero'] = df_valid_esc['Genero'].replace({'Masculino': 0, 'Feminino': 1})
df_valid_esc['Ano_escolar'] = df_valid_esc['Ano_escolar'].replace({
    'EMedio': 9,
    'Nono': 8, 
    'Oitavo': 7,  
    'Sétimo': 6, 
    'Sexto': 5, 
    'Quinto': 4, 
    'Quarto': 3, 
    'Terceiro': 2,
    'Segundo': 1, 
    'Primeiro': 0})

# Leitura
df_train_leit = df_train_leit.replace({'Sim': 1, 'Não': 0})
df_train_leit['TDE_MG_Esc'] = df_train_leit['TDE_MG_Esc'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})

df_valid_leit = df_valid_leit.replace({'Sim': 1, 'Não': 0})
df_valid_leit['TDE_MG_Esc'] = df_train_leit['TDE_MG_Esc'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})

df_train_leit
```

```python
print(df_train_arit.shape)
print(df_train_esc.shape)
print(df_train_leit.shape)
print(df_valid_arit.shape)
print(df_valid_esc.shape)
print(df_valid_leit.shape)

df_train_arit.to_csv("databases/AG/Aritmética/TreinamentoDesbalanceadoAritPreprocessada.csv", sep=";", index=False)
df_train_esc.to_csv("databases/AG/Escrita/TreinamentoDesbalanceadoEscPreprocessada.csv", sep=";", index=False)
df_train_leit.to_csv("databases/AG/Leitura/TreinamentoDesbalanceadoLeitPreprocessada.csv", sep=";", index=False)

df_valid_arit.to_csv("databases/AG/Aritmética/TesteAritPreprocessada.csv", sep=";", index=False)
df_valid_esc.to_csv("databases/AG/Escrita/TesteEscPreprocessada.csv", sep=";", index=False)
df_valid_leit.to_csv("databases/AG/Leitura/TesteLeitPreprocessada.csv", sep=";", index=False)
```

## Tranformacoes base AG + Random + Especialista

```python
base_path = "databases/AG_Random_Especialista"

# Load originaal datasets
df_train_arit = pd.read_csv(base_path + '/Aritmética/TreinamentoDesbalanceadoArit.csv', sep=";", encoding="latin1")
df_train_esc = pd.read_csv(base_path + "/Escrita/TreinamentoDesbalanceadoEsc.csv", sep=";", encoding="latin1")
df_train_leit = pd.read_csv(base_path + "/Leitura/TreinamentoDesbalanceadoLeit.csv", sep=";", encoding="latin1")

df_test_arit = pd.read_csv(base_path + "/Aritmética/TesteArit.csv", sep=";", encoding="latin1")
df_test_esc = pd.read_csv(base_path + "/Escrita/TesteEsc.csv", sep=";", encoding="latin1")
df_test_leit = pd.read_csv(base_path + "/Leitura/TesteLeit.csv", sep=";", encoding="latin1")

df_train_esc
```

```python
def preprocess_data(df, target_name, keep_columns):
    preprocesed_data: pd.DataFrame = df.drop_duplicates(ignore_index=True)
    print('Removing duplicate instances...')
    print(f'Before: {df.index.size} After: {preprocesed_data.index.size}')

    target_data = preprocesed_data[target_name]
    preprocesed_data = preprocesed_data.drop(columns=[target_name])

    keep_data = preprocesed_data[keep_columns]
    preprocesed_data = preprocesed_data.drop(columns=keep_columns)

    columns = preprocesed_data.columns.to_numpy()
    print(f"Num. columns: {columns.size}")

    preprocesed_data: np.ndarray = preprocesed_data.to_numpy()
    encoder = OrdinalEncoder()
    preprocesed_data = encoder.fit_transform(preprocesed_data)
    preprocesed_data = preprocesed_data.astype(np.int32)

    print('Catagories: ', end="")
    for categorie in encoder.categories_:
        print(categorie,end="  ")
    
    print()

    preprocesed_data: pd.DataFrame = pd.DataFrame(preprocesed_data, columns=columns)
    preprocesed_data = preprocesed_data.join(keep_data)
    preprocesed_data = preprocesed_data.join(target_data)
    
    return preprocesed_data
    
    



    
keep_columns = ['Idade', 'Altura', 'Quantidade_irmaos', 'Indulgencia', 'Acolhimento_emocional', 'Dialogo', 'Autonomia', 'Coercao_fisica', 'Hostilidade_verbal', 'Punicao', 'Hiperatividade_mae']
print(len(keep_columns))
preprocess_data(df_train_arit, 'TDE_MG_Arit', keep_columns)
```

```python
# aritmetica
keep_columns = ['Idade', 'Altura', 'Quantidade_irmaos', 'Indulgencia', 'Acolhimento_emocional', 'Dialogo', 'Autonomia', 'Coercao_fisica', 'Hostilidade_verbal', 'Punicao', 'Hiperatividade_mae']
print('Preprocessando Aritimetica Treino')
preprocess_data(df_train_arit, 'TDE_MG_Arit', keep_columns).to_csv(base_path + "/Aritmética/TreinamentoAritPrep.csv", index=False, sep=';')
print()
print('Preprocessando Aritimetica Teste')
preprocess_data(df_test_arit, 'TDE_MG_Arit', keep_columns).to_csv(base_path + "/Aritmética/TesteAritPrep.csv", index=False, sep=';')
print()

# escrita
keep_columns = ['Idade', 'Quantidade_irmaos', 'Parceiros_mae', 'Idade_mae', 'Indulgencia', 'Acolhimento_emocional', 'Dialogo', 'Autonomia', 'Coercao_fisica', 'Hostilidade_verbal', 'Punicao', 'Desatencao_mae' ]
print('Preprocessando Escrita Treino')
preprocess_data(df_train_esc, 'TDE_MG_Esc', keep_columns).to_csv(base_path + "/Escrita/TreinamentoEscPrep.csv", index=False, sep=';')
print()
print('Preprocessando Escrita Teste')
preprocess_data(df_test_esc, 'TDE_MG_Esc', keep_columns).to_csv(base_path + "/Escrita/TesteEscPrep.csv", index=False, sep=';')
print()

# Leitura
keep_columns = ['Idade', 'Indulgencia', 'Acolhimento_emocional', 'Dialogo', 'Autonomia', 'Coercao_fisica', 'Hostilidade_verbal', 'Punicao']
print('Preprocessando Leitura Treino')
preprocess_data(df_train_leit, 'TDE_MG_Leit', keep_columns).to_csv(base_path + "/Escrita/TreinamentoLeitPrep.csv", index=False, sep=';')
print()
print('Preprocessando Leitura Teste')
preprocess_data(df_test_leit, 'TDE_MG_Leit', keep_columns).to_csv(base_path + "/Escrita/TesteLeitPrep.csv", index=False, sep=';')
print()

```
