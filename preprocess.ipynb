```python
import pandas as pd

# Load originaal datasets
df_train_arit = pd.read_csv("databases/AG/Aritmética/TreinamentoDesbalanceadoArit.csv", sep=";", encoding="latin1")
df_train_esc = pd.read_csv("databases/AG/Escrita/TreinamentoDesbalanceadoEsc.csv", sep=";", encoding="latin1")
df_train_leit = pd.read_csv("databases/AG/Leitura/TreinamentoDesbalanceadoLeit.csv", sep=";", encoding="latin1")

df_valid_arit = pd.read_csv("databases/AG/Aritmética/TesteArit.csv", sep=";", encoding="latin1")
df_valid_esc = pd.read_csv("databases/AG/Escrita/TesteEsc.csv", sep=";", encoding="latin1")
df_valid_leit = pd.read_csv("databases/AG/Leitura/TesteLeit.csv", sep=";", encoding="latin1")

df_train_leit
```

```python
# Aritmetica
df_train_arit = df_train_arit.replace({'Sim': 1, 'Não': 0})
df_train_arit['TDE_MG_Esc'] = df_train_arit['TDE_MG_Esc'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_train_arit['Escolaridade_mae'] = df_train_arit['Escolaridade_mae'].replace({'EFundamental': 0, 'EMedio': 1, 'ESuperior':2})
df_train_arit['Parto'] = df_train_arit['Parto'].replace({'Normal': 0, 'Cesária': 1})

df_valid_arit = df_valid_arit.replace({'Sim': 1, 'Não': 0})
df_valid_arit['TDE_MG_Esc'] = df_valid_arit['TDE_MG_Esc'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_valid_arit['Escolaridade_mae'] = df_valid_arit['Escolaridade_mae'].replace({'EFundamental': 0, 'EMedio': 1, 'ESuperior':2})
df_valid_arit['Parto'] = df_valid_arit['Parto'].replace({'Normal': 0, 'Cesária': 1})

# Escrita
df_train_esc = df_train_esc.replace({'Sim': 1, 'Não': 0})
df_train_esc['TDE_MG_Leit'] = df_train_esc['TDE_MG_Leit'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_train_esc['TDE_MG_Arit'] = df_train_esc['TDE_MG_Arit'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_train_esc['Genero'] = df_train_esc['Genero'].replace({'Masculino': 0, 'Feminino': 1})
df_train_esc['Ano_escolar'] = df_train_esc['Ano_escolar'].replace({
    'EMedio': 9,
    'Nono': 8, 
    'Oitavo': 7,  
    'Sétimo': 6, 
    'Sexto': 5, 
    'Quinto': 4, 
    'Quarto': 3, 
    'Terceiro': 2,
    'Segundo': 1, 
    'Primeiro': 0})


df_valid_esc = df_valid_esc.replace({'Sim': 1, 'Não': 0})
df_valid_esc['TDE_MG_Leit'] = df_valid_esc['TDE_MG_Leit'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_valid_esc['TDE_MG_Arit'] = df_valid_esc['TDE_MG_Arit'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})
df_valid_esc['Genero'] = df_valid_esc['Genero'].replace({'Masculino': 0, 'Feminino': 1})
df_valid_esc['Ano_escolar'] = df_valid_esc['Ano_escolar'].replace({
    'EMedio': 9,
    'Nono': 8, 
    'Oitavo': 7,  
    'Sétimo': 6, 
    'Sexto': 5, 
    'Quinto': 4, 
    'Quarto': 3, 
    'Terceiro': 2,
    'Segundo': 1, 
    'Primeiro': 0})

# Leitura
df_train_leit = df_train_leit.replace({'Sim': 1, 'Não': 0})
df_train_leit['TDE_MG_Esc'] = df_train_leit['TDE_MG_Esc'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})

df_valid_leit = df_valid_leit.replace({'Sim': 1, 'Não': 0})
df_valid_leit['TDE_MG_Esc'] = df_train_leit['TDE_MG_Esc'].replace({'Inferior': 0, 'Médio': 1, 'Superior':2})

df_train_leit
```

```python
print(df_train_arit.shape)
print(df_train_esc.shape)
print(df_train_leit.shape)
print(df_valid_arit.shape)
print(df_valid_esc.shape)
print(df_valid_leit.shape)

df_train_arit.to_csv("databases/AG/Aritmética/TreinamentoDesbalanceadoAritPreprocessada.csv", sep=";", index=False)
df_train_esc.to_csv("databases/AG/Escrita/TreinamentoDesbalanceadoEscPreprocessada.csv", sep=";", index=False)
df_train_leit.to_csv("databases/AG/Leitura/TreinamentoDesbalanceadoLeitPreprocessada.csv", sep=";", index=False)

df_valid_arit.to_csv("databases/AG/Aritmética/TesteAritPreprocessada.csv", sep=";", index=False)
df_valid_esc.to_csv("databases/AG/Escrita/TesteEscPreprocessada.csv", sep=";", index=False)
df_valid_leit.to_csv("databases/AG/Leitura/TesteLeitPreprocessada.csv", sep=";", index=False)
```
